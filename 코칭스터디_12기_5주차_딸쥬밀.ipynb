```python
def bs31(num=None):
  import random
  present_num = 0

  print('베스킨라빈스 써리원 게임')
  print('-'*17)
  # 현재 숫자가 31보다 작을 경우 반복문 실행
  while present_num < 31:
    game = True
    my = input('My turn - 숫자를 입력하세요.:')
    my = my.split()

    # my 입력을 정수형으로 변경
    my_int = []
    for i in my:
      my_int.append(int(i))

      # 31이 입력될 경우 게임 종료
      if i == 31:
        game = False
        break
    if game == False:
      break

    # my 마지막 숫자를 현재 숫자로 할당
    present_num = my_int[-1]
    print(f'현재 숫자 : {present_num}')

    # 컴퓨터가 외칠 숫자 개수 설정
    computer_turn_num = random.randint(1, 3)

    # 컴퓨터가 외치는 숫자 설정
    computer_int = list(range(present_num+1, present_num+1+computer_turn_num))
    present_num = computer_int[-1]

    # 31보다 작을 경우 그대로 출력
    if present_num <= 31:
      print('컴퓨터 :', *computer_int, sep=' ')
      print(f'현재 숫자 : {present_num}\n')
    # 31이거나 클 경우 게임 종료
    else:
      for i in computer_int:
        if i <= 31:
          print('컴퓨터 :', *computer_int, sep=' ')
        else:
          game = False
    if game == False:
      break
```
