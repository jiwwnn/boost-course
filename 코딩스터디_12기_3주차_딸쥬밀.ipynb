```python
def gugudan(num):
  print('{} 단'.format(num))
  for i in range(1, 10, 2):
    if i*8 <= 50:
      print(num, 'X', i, '=', 8*i)
```

```python
number = int(input('몇 단? :'))
gugudan(number)
```

```python
def find_even_number(n,m):
  numbers = [i for i in range(n, m+1)]
  index = len(numbers) // 2
  median = numbers[index]
  for i in numbers:
    if i % 2 == 0 and i != median:
      print(i, '짝수')
    elif i % 2 == 0 and i == median:
      print(i, '중앙값')
```

```python
n = int(input('첫 번째 수 입력: '))
m = int(input('두 번째 수 입력: '))
find_even_number(n,m)
```
