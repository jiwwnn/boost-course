#### 1) Syntax Error

```python
for i in range(10)
  print(i)
```

```python
# 디버깅 : 반복문 문법에 맞게 ':' 추가
for i in range(10):
  print(i)
```

```python
a = 'hi'

if a = 'hi':
  print('co-ha')
```

```python
# 디버깅 : 값이 동일함을 나타내기 위해 '=' 대신 '==' 사용
a = 'hi'

if a == 'hi':
  print('co-ha')
```

#### 2) Value Error

```python
a = int('hello')
a
```

```python
# 디버깅 : 문자형은 정수형으로 변형이 안되므로 숫자값으로 변경
a = int(12.23)
a
```

```python
a = []
min(a)
```

```python
# 디버깅 : 빈 리스트가 아닌 값이 존재하는 리스트로 변경
a = [1, 2, 3, 4, 5]
min(a)
```

#### 3) Type Error

```python
a = 123
a[2]
```

```python
# 디버깅 : 정수형은 슬라이싱할 수 없으므로 리스트나 문자형으로 변경
a = [1, 2, 3]
a[2]
```

```python
for i in range(10.5):
  print(i)
```

```python
# 디버깅 : 반복문에서는 range 안에 실수값이 아닌 정수값 입력
for i in range(10):
  print(i)
```
