```python
# 이름, 실적
member_names = ["갑돌이", "갑순이", "을돌이", "을순이", "병돌이", "병순이"]
member_records = [[4,5,3,5,6,5,3,4,1,3,4,5],[2,3,4,3,1,2,0,3,2,5,7,2],
           [1,3,0,3,3,4,5,6,7,2,2,1],[3,2,9,2,3,5,6,6,4,6,9,9],
           [8,7,7,5,6,7,5,8,8,6,10,9],[7,8,4,9,5,10,3,3,2,2,1,3]]
```

```python
def sales_management(member_names, member_records):
  import numpy as np

  sales_dict = {}
  for n, r in zip(member_names, member_records):
    sales_dict[n] = np.mean(r)

  sales_dict = dict(sorted(sales_dict.items(), key=lambda x:x[1], reverse=True))
  sales_lst = list(sales_dict.keys())

  for k, v in sales_dict.items():
    if v > 5 and k in sales_lst[:2]:
      print('보너스 대상자', k)
    elif v < 3 and k in sales_lst[5:]:
      print('\n면담 대상자', k)
```

```python
sales_management(member_names, member_records)
```

```python
# 종목 이름, 수량, 매수 평균 금액
stocks = "삼성전자/10/85000,카카오/15/130000,LG화학/3/820000,NAVER/5/420000"

# 종목별 판매가
sells = [82000, 160000, 835000, 410000]
```

```python
# cf.수익률 = (매도가격 - 매수가격)/매수가격

def stock_profit(stocks, sells):
  rate_dic = {}
  for stock, sell in zip(stocks.split(','), sells):
    name, buy = stock.split('/')[0], int(stock.split('/')[2])

    rate = (sell - buy)/buy
    rate_dic[name] = rate*100

  rate_dic = dict(sorted(rate_dic.items(), key=lambda x:x[1], reverse=True))
  rate_dic
  for i in rate_dic.items():
    name, rate = i[0], i[1]
    print(f'{name}의 수익률 {rate:.3}')
```

```python
stock_profit(stocks, sells)
```
